```python
import pandas as pd

notebook_dir = Path.cwd()
INPUT_FILE = notebook_dir.parent / "data/raw/ratings.csv"
OUTPUT_FILE = notebook_dir.parent / "data/raw/ratings_reduced.csv"

CHUNK_SIZE = 1_000_000      # rows per chunk
FRACTION = 0.15             # keep 15% of each chunk (~<512MB in practice)

chunks = pd.read_csv(
    INPUT_FILE,
    chunksize=CHUNK_SIZE
)

reduced_chunks = []

for chunk in chunks:
    reduced_chunks.append(
        chunk.sample(frac=FRACTION, random_state=42)
    )

reduced_ratings = pd.concat(reduced_chunks)

reduced_ratings.to_csv(
    OUTPUT_FILE,
    index=False
)

print("Reduced file created:", reduced_ratings.shape)
```
